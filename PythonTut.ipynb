```python
#NESTED LOOPS
```

```python
for i in range(3):     # outer loop takes values 0,1,2
    for j in range(2): #inner loop take values 0,1
        print(i, j)    # print the current pair
        
                       # i = 0 → inner loop runs: j = 0, then j = 1
                       # i = 1 → inner loop runs: j = 0, then j = 1
                       # i = 2 → inner loop runs: j = 0, then j = 1
```

mini challenge
Write a program that: 
1. Prints numbers from 1 to 10 
2. Prints only even numbers 
3. Calculates the total sum 

```python
# Task 1: Print numbers from 1 to 10
for i in range(1, 11):
    print(i)
print("\n")
# Task 2: Print only even numbers (from 1 to 10)
for i in range(1, 11):
    if i % 2 == 0:  # Check if number is even
        print(i)
print("\n")
# Task 3: Calculate the total sum of numbers from 1 to 10
totalSum = 0
for i in range(1, 11):
    total_sum += i  # Add each number to the running total
print(totalSum)
```

```python
# Password Attempts 
```

```python
correct_password = "ICS4U"
attempts = 0

while attempts < 3:
    user_input = input("Enter password: ")
    
    if user_input == correct_password:
        print("Access granted.")
        break
    else:
        print("Incorrect password.")
        attempts += 1

if attempts == 3:
    print("Account locked.")
```

```python
#simple Grade calculator
```

```python
while True: 
   score = int(input("Enter score (or -1 to quit): ")) 
    
   if score == -1: 
      break 
    
   if score >= 75: 
      print("A") 
   elif score >= 70: 
      print("B") 
   elif score >= 60: 
      print("C") 
   else: 
      print("E")
```
